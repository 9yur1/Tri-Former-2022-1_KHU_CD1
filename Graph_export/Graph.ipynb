```python
### S,T,FR (Triformer)
import numpy as np
trend = np.load('./Triformer/trend.npy')
seasonal = np.load('./Triformer/seasonal.npy')
noise = np.load('./Triformer/noise.npy')
pred = np.load('./Triformer/pred.npy')

import matplotlib.pyplot as plt
%matplotlib inline
y_trend=[]
y_seasonal=[]
y_noise=[]
y_pred=[]

for i in range(96):
        y_trend.append(trend[0][0][i][7])      

for i in range(96):
        y_seasonal.append(seasonal[0][0][i][7])
        
for i in range(96):
        y_noise.append(noise[0][0][i][7])
        
for i in range(96):
        y_pred.append(pred[0][0][i][7])

   
plt.figure(figsize=(20,40))
plt.subplot(811)
plt.title("Trend")
plt.plot(y_trend)
plt.figure(figsize=(20,5))
plt.subplot(812)
plt.title("Seasonal")
plt.plot(y_seasonal)
plt.figure(figsize=(20,20))
plt.subplot(813)
plt.title("Noise")
plt.plot(y_noise)
plt.figure(figsize=(20,40))
plt.subplot(814)
plt.title("Prediction")
plt.plot(y_pred)
```
